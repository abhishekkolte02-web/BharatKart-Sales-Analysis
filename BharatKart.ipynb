```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('bharatkart_sales_dataset.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
df=df.drop('company_name',axis=1)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()  # to check missing values in data
```

```python
df['city'] = df['city'].fillna('Unknown') # to fill missing values as Unknown for missing city
df['state'] = df['state'].fillna('NA')  # to fill missing states as NA
df['product_category'] = df['product_category'].fillna('Unknown') # to fill missing product_category as Unknown
df['product_name'] = df['product_name'].fillna('Unknown')
```

```python
df['city'].unique()
```

```python
df['city'] = df['city'].replace('unknown','Unknown')
```

```python
df['city'].unique()
```

```python
df['state'].unique()
```

```python
df['discount_percent'] = df['discount_percent'].fillna(0)
```

```python
df["rating"] = df["rating"].fillna(df.groupby('product_category')['rating'].transform('median')) # To fill missing values by respective product categories
```

```python
df['payment_mode'] = df['payment_mode'].fillna('Unkonwn')
```

```python
df.isnull().sum()
```

```python
df["customer_name"] = df["customer_name"].fillna('Unkonwn')
```

```python
df["customer_name"] = df["customer_name"].str.title()
```

```python
df['order_value'] = df['order_value'].fillna(df.groupby('product_category')['order_value'].transform('median'))
```

```python
df.info()
```

```python
df['rating'] = df['rating'].astype(int)
```

```python
df['order_date'] = pd.to_datetime(df['order_date'])
```

```python
df['month'] = df['order_date'].dt.month
df['year'] = df['order_date'].dt.year
```

```python
df.to_csv("bharatkart_cleaned.csv", index = False)
```
